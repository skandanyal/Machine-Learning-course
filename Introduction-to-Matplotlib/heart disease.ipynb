```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

```python
# distribution of ages
# heart_disease.plot(x='age', kind='histogram')
%matplotlib inline
heart_disease['age'].plot.hist(bins=20, color='orange', edgecolor='black');
plt.show()
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots=True); # not very efficient
plt.tight_layout()
plt.show()
```

## Which one should you use? (call pyplot or use OO method)

* When plotting smtg quickly, use pyplot
* When plotting something advanced, use OO method

## Working with data of patients above the age of 50

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] >= 50]
over_50
```

```python
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
plt.show()
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c=['orange' if b else 'blue' for b in over_50['target']],
            )
plt.show()
```

```python
# oo method 
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c=['orange' if b else 'green' for b in over_50['target']],
            ax=ax);
```
