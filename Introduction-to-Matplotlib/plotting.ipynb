## Making figures using NumPy arrays

```python
# data 
x = np.linspace(0,10,101)
```

```python
# plot the data
#%matplotlib inline
fig, ax = plt.subplots()
ax.plot(x, x**0.5);
plt.show() #creates a linr plot
```

```python
# a scatterplot
fig, ax = plt.subplots()
ax.scatter(x, x**2)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# a plot from a dictionary
%matplotlib inline
nut_butter_prices = {'Almond butter': 10,
                     'Peanut butter': 8,
                     'Cashew butter': 12}  
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
# plt.title("Prices of different kinds of butter ($)")
# plt.xlabel("Varieties of butters")
# plt.ylabel("Cost ($)")
ax.set(title='Prices of different kinds of butter ($)',
      ylabel='Cost ($)',
      xlabel='Varieties of butters');
plt.show()
```

```python
fig, ax = plt.subplots()
# ax.barh(nut_butter_prices.keys(), nut_butter_prices.values)
# need to make the inputs as lists
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
# making normal data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
plt.show()
```

```python
# subplot opt1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(8,5))

ax1.plot(x,x*3)
ax1.set(title="line graph")

ax2.scatter(y,y)
ax2.set(title="scatter plot")

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(title="bar graph")

ax4.hist(x)
ax4.set(title="histogram")

plt.tight_layout()
plt.show()
```

```python
# subplot opt2
y = [i for i in range(10)]
fig, ax = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(8,5))

ax[0,0].plot(x,x*3)
ax[0,0].set(title="line graph")

ax[0,1].scatter(y,np.sin(y))
ax[0,1].set(title="scatter plot")

ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,0].set(title="bar graph")

ax[1,1].hist(x, color='orange', edgecolor='black')
ax[1,1].set(title="histogram")

plt.tight_layout()
plt.show()
```

## Plotting from Pandas dataframe

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# trial ke liye
%matplotlib inline

ts = pd.Series(np.random.random(1000),
              index = pd.date_range("1/1/2024", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '', regex=True).astype(int)
```

```python
car_sales
```

```python
# didnt remove the last two zeros from the prices column
```

```python
car_sales['Sales Date'] = pd.date_range('1/1/2025', periods=len(car_sales))
car_sales
```

```python
type(car_sales['Price'][0])
```

```python
car_sales['Total sales ($)'] = car_sales['Price'].cumsum()
car_sales
```

```python
# plotting car sales
car_sales.plot(x="Sales Date", 
               y="Total sales ($)");
```

```python
plt.show()
```
